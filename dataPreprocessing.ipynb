```python
import pandas as pd
```

```python
df = pd.read_csv('amazon_reviews_us_Grocery_v1_00.tsv', sep='\t', header=0, on_bad_lines='skip')

print(df.head())
print(df.shape)

## 9079 missing lines, which is 0.4%
```

```python
"""              FILTER ANY DATA NOT IN US, NOT IN GROCERY OR NOT VERIFIED                 """

""" Explain why we don't want unverified reviews if we decide this """
```

```python
num_not_us = df.shape[0] - df.marketplace.value_counts()['US']
print(num_not_us)

## All data is US marketplace
```

```python
num_not_grocery = df.shape[0] - df.product_category.value_counts()['Grocery']
print(num_not_grocery)

df = df[df.product_category == 'Grocery']
print(df.shape)
```

```python
verified_count = df.verified_purchase.value_counts()['Y']
percent_verified = 100*verified_count/df.shape[0]
print(percent_verified)

## 82% verified

df = df[df.verified_purchase == 'Y']
```

```python
""" Reduce dataframe to just the columns we might use """
```

```python
df.drop(columns=['marketplace', 'customer_id', 'review_id', 'product_category', 'vine', 'verified_purchase'],inplace=True)
print(df.columns)
```
